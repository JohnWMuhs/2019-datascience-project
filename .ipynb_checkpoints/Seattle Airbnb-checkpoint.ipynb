```python
# your code goes here
import gmaps
import gmaps.datasets
from matplotlib.cm import viridis
from matplotlib.colors import to_hex

gmaps.configure(api_key="AIzaSyAUad_1plUmJaCo0ZiVVULjOYZ8u-85WkA") # Your Google API key


# Scale the Sales Price values between 0 and 1
min_price = np.min(df['price'])
max_price = np.max(df['price'])
price_range = max_price - min_price
print(price_range)

# Calculate color and convert to hex
def calculate_color(price):
    """
    Convert the GINI coefficient to a color
    """
    # make gini a number between 0 and 1
    normalized_price = (price - min_price) / price_range

    # invert gini so that high inequality gives dark color
    inverse_price = 1.0 - normalized_price

    # transform the gini coefficient to a matplotlib color
    mpl_color = viridis(inverse_price)

    # transform from a matplotlib color to a valid CSS color
    gmaps_color = to_hex(mpl_color, keep_alpha=False)

    return gmaps_color

colors = []
for price in df['price']:
    color = calculate_color(price)
    colors.append(color)

locations = df[['latitude', 'longitude']]
weights = df['price']

fig = gmaps.figure()

fig.add_layer(gmaps.heatmap_layer(locations,weights = weights))
#fig.add_layer(gmaps.symbol_layer(locations,fill_color=colors,
#    stroke_color=colors,
#    fill_opacity=0.8))
fig
```
