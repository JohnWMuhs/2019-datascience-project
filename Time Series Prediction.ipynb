# ARIMA Time Series Prediction

In this notebook, we will use atime-series prediction technique to train and predict CAISO energy consumption data without any external variables. We will use a technique called ARIMA (Auto Regressive Integration Moving Average) prediction technique. This technique takes into account trends, seasonality, and residuals to model a future prediction within a certain confidence interval.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib.pylab import rcParams
from scipy import stats
rcParams['figure.figsize'] = 18,6
```

## Import CAISO Data and Split into Train/Test Sets

```python
df = pd.read_csv('energydata.csv', index_col = 'DateTime', parse_dates=True, usecols = [0,1])
print(df.quantile(0.01))
df = df.where(df > df.quantile(0.01),np.nan)
df = df.fillna(method='ffill')
dfTrain = df[df.index < pd.to_datetime('January 1 2018').tz_localize('US/Pacific')].dropna()
dfTest = df[df.index >= pd.to_datetime('January 1 2018').tz_localize('US/Pacific')].dropna()
dfTest = dfTest[dfTest.index < pd.to_datetime('January 1 2019').tz_localize('US/Pacific')].dropna()
```

## Test Dataset for Stationarity

```python
from statsmodels.tsa.stattools import adfuller
def test_stationary(timeseries):
    timewindow = 20
    
    # Calculate Rolling Statistics
    moving_avg = timeseries.rolling(window=timewindow).mean()
    moving_std = timeseries.rolling(window=timewindow).std()
    
    # Plot Rolling Statistics with Raw Data
    plt.plot(timeseries,label='Raw Data')
    plt.plot(moving_avg,label='Rolling Avg ({})'.format(timewindow))
    plt.plot(moving_std,label='Rolling Std ({})'.format(timewindow))
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    
    #Perform Dicky Fuller Test
    print('Results of Dicky Fuller Test')
    dftest = adfuller(timeseries['CAISO_HourlyLoad'],autolag = 'AIC')
    dfoutput = pd.Series(dftest[0:4],index=['Test Statistic:','p-Value:','#Lags Used','# Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    
```

```python
test_stationary(dfTrain.diff().fillna(0))
```

## Check ACF and PACF Plots

Add code here from [youtube video](https://www.youtube.com/watch?v=e8Yw4alG16Q) at approximate 29 minutes.

```python
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(dfTrainDiff, nlags=20)
lag_pacf = pacf(dfTrainDiff, nlags=20, method='ols')

#Plot ACF
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfTrainDiff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(dfTrainDiff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfTrainDiff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(dfTrainDiff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

## Fit ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA

#AR Model
model = ARIMA(dfTrainDiff,order = (0, 1, 2))
resultsAR = model.fit(disp=-1)
plt.plot(dfTrainDiff)
plt.plot(resultsAR.fittedvalues)
plt.title('RSS %.4f'% sum((resultsAR.fittedvalues-dfTrainDiff['CAISO_HourlyLoad'])**2))
print('Plotting AR Model')
```

```python
#MA Model
model = ARIMA(dfTrainDiff,order = (0, 1, 2))
resultsMA = model.fit(disp=-1)
plt.plot(dfTrainDiff)
plt.plot(resultsMA.fittedvalues)
plt.title('RSS %.4f'% sum((resultsMA.fittedvalues-dfTrainDiff['CAISO_HourlyLoad'])**2))
print('Plotting MA Model')
```

```python
#ARIMA Model
model = ARIMA(dfTrainDiff,order = (0, 1, 2))
resultsARIMA = model.fit(disp=-1)
plt.plot(dfTrainDiff)
plt.plot(resultsARIMA.fittedvalues)
plt.title('RSS %.4f'% sum((resultsARIMA.fittedvalues-dfTrainDiff['CAISO_HourlyLoad'])**2))
print('Plotting ARIMA Model')
```

```python
dfPredictDiff = pd.Series(resultsARIMA.fittedvalues, copy=True)
print(dfPredictDiff.head())
```

```python
dfPredictCumSum = dfPredictDiff.cumsum().add(dfTrain['CAISO_HourlyLoad'].iloc[0])
dfPredict = pd.Series(dfTrain['CAISO_HourlyLoad'].iloc[0],index=dfTrain.index)
dfPredict = dfPredict.add(dfPredictCumSum,fill_value=0)
dfPredict.head()
```

```python
plt.plot(dfTrain)
plt.plot(dfPredict)
```
